```python
import pandas as pd
#import uuid
from sqlalchemy import create_engine
from getpass import getpass
```

```python
user = input('videosurvey DB user name:')
password = getpass('videosurvey DB password:')
engine = create_engine(f'mysql+pymysql://{user}:{password}@mysql.guaminsects.net/videosurvey')
engine.connect() # test connection
```

# videos table

```python
# Read video table from videosurvey DB

connection = engine.connect()
df = pd.read_sql_table('video', engine)
df
```

```python
df_videos = df
df_videos.rename(columns={'Video File':'video_id'}, inplace=True)

# Retain only rows which have video_id in a format like 20200630_131234.mp4
regex = r"2020\d\d\d\d_\d\d\d\d\d\d.mp4"
df_videos = df_videos[df_videos.video_id.str.match(regex)]    

df_videos
```

```python
df_videos.to_sql('videos', connection, if_exists='append', index=False)
```

# frames table

```python
# Read objects table from videosurvey DB

connection = engine.connect()
df_objects = pd.read_sql_table('objects', engine)

regex = r"2020\d\d\d\d_\d\d\d\d\d\d.mp4"
df_objects = df_objects[df_objects.video_file.str.match(regex)]          # dumps rows with invalid video_file

df_objects
```

```python
df_frames = df_objects
df_frames.rename(columns={'video_file':'video_id'}, inplace=True)
df_frames = df_frames.groupby(['video_id','frame']).first()         # gets one row for each frame
df_frames = df_frames.reset_index()                                   # converts indices to ordinary columns
df_frames['frame_id'] = df_frames.apply(lambda r: f"{r.video_id}-{r.frame}", axis=1)
df_frames = df_frames[['frame_id', 'video_id', 'frame','timestamp','lat','lon']] # dumps columns we don't need
df_frames
```

```python
df_frames.to_sql('frames', connection, if_exists='append', index=False)
```
