# Workaround

https://stackoverflow.com/questions/42672807/insert-geometry-point-mysql-from-pandas-dataframe

A work around is to create a temporary table with the String of the geometrical information that need to be added and then update the point_test table with a call to ST_GeomFromText from the temporary table.

Assuming database with table point_test with id (VARCHAR(5)) and coords(POINT):

a.Create dataframe df as an example with point "A" and "B"

```python
import numpy as np

dfd = np.array([['id','geomText'],
            ["A","POINT( 50.2 5.6 )"],
            ["B","POINT( 50.2 50.4 )"]])
df=pd.DataFrame(data=dfd[1:,:], columns=dfd[0,:])
```

b.Add point "A" and "B" into point_test but only the id and add the string "geomText" into the table temp_point_test

```python
df[['id']].to_sql("point_test",engine, index=False, if_exists="append")
df[['id', 'geomText']].to_sql("temp_point_test",engine, index=False, if_exists="append")
```

c. Update table point_test with the point from table temp_point_test applying the ST_GeomFromText() to the select. Finally, drop temp_point_test:

```python
conn = engine.connect()
```

```python
conn.execute('ALTER TABLE point_test ADD coords POINT')
```

```python
conn.execute("update point_test pt set pt.coords=(select ST_GeomFromText(geomText) from temp_point_test tpt "+
                     "where pt.id=tpt.id)")
```

```python
conn.execute("drop table temp_point_test")   
```

```python
conn.close()
```

# My Workaround

```python
# Add a CSV to the MySQL objects table
# Includes a workaround to create geometry column (coords) so that QGIS can plot points on a map

import pandas as pd
from sqlalchemy import create_engine

CSV = '/media/aubrey/9C33-6BBD/20200630_132859/20200630_132859_gps_obs.csv'
VIDEO = '20200630_132859.mp4'
USER = 'savan'
PASSWORD = 'oryctes_rhinoceros'

engine = create_engine(f'mysql+pymysql://{USER}:{PASSWORD}@mysql.guaminsects.net/videosurvey')

df = pd.read_csv(CSV)
df['video_file'] = VIDEO
conn = engine.connect()
conn.execute('DROP TABLE IF EXISTS temp;')
df.to_sql("temp", engine, index=False, if_exists="replace")

conn.execute('ALTER TABLE temp ADD coords POINT;')
conn.execute('UPDATE temp SET coords=POINT(lon,lat);')
conn.execute('INSERT INTO objects SELECT * FROM temp;')
conn.close()
```
