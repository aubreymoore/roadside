```python
df = pd.read_csv('20200801.csv')
df = df[pd.notnull(df['annotation'])] # drop rows where annotation is NaN
df_gcp = df[['annotation','lon','lat','elevation']]
df_gcp.to_csv('gcp_lat_lon.csv', index=False)
df_gcp
```

```python
df_gcp_pixel_loc = pd.read_csv('gcp_pixel_loc.csv')
df_gcp_pixel_loc
```

```python
df_gcp_lat_lon = pd.read_csv('gcp_lat_lon.csv')
df_gcp_lat_lon
```

```python
# pd.merge(left, right, how='inner', on=None, left_on=None, right_on=None,
#          left_index=False, right_index=False, sort=True,
#          suffixes=('_x', '_y'), copy=True, indicator=False,
#          validate=None)

# x y z pixelx pixely imagename

df_merged = pd.merge(left=df_gcp_pixel_loc, right=df_gcp_lat_lon)
df_merged
```

```python
s = 'WGS84\n'
for _, r in df_merged.iterrows():
    s += f'{r.lon} {r.lat} {r.elevation:.0f} {r.X} {r.Y} {r.image}\n'
print(s)
with open('gcp_list.txt', 'w') as f:
    f.write(s)
```
