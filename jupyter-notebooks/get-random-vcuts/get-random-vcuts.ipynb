```python
def str2int(s):
    return int(float(s))

# str2int('7.53')
```

```python
def polygon2bb(row):
    points = row['points'].split(';')
    xlist = []
    ylist = []
    for point in points:
        point = point.split(',')
        xlist.append(str2int(point[0]))
        ylist.append(str2int(point[1]))
    row['xbr'] = max(xlist)
    row['xtl'] = min(xlist)
    row['ybr'] = max(ylist)
    row['ytl'] = min(ylist)
    return row
```

```python
def get_bounding_boxes(xml_file):
    """
    Returns a pandas dataframe.
    """   
    tree = ET.parse(xml_file)
    root = tree.getroot()
    
    print(ET.tostring(root.find('meta'), encoding='unicode'))

    mylist = []
    for track in root.findall('track'):
        for polygon in track.findall('polygon'):
            mydict = polygon.attrib
            mydict.update(track.attrib)
            mylist.append(mydict)
    df = pd.DataFrame(mylist)
    df = df[(df.occluded=='0') & (df.outside=='0')]
    df.drop(['occluded', 'outside'], axis=1, inplace=True)
    df.rename({'id': 'track_id'}, axis=1, inplace=True)
    df = df.apply(polygon2bb, axis=1)
    df.frame = df.frame.apply(lambda x: str2int(x))
    df.track_id = df.track_id.apply(lambda x: str2int(x))
    return df

# df = get_bounding_boxes(f'{DATADIR}/{CVATXMLFILE}')
# df
```

```python
def tag_false_positives():
    """
    Simple way to tag images inspired by:
    https://towardsdatascience.com/label-your-images-easily-using-this-jupyter-notebook-code-4102037b7821    
    """
    imagelist = glob.glob('*.jpg')
    labellist = []
    for image in imagelist:
        display(Image(image)) 
        inp = input(f'{image}\nEnter x if this is a false positive; leave blank otherwise:')
        if inp=='x':
            labellist.append('false positive')
            print('false positive')
        else:
            labellist.append('ok')
            print('ok')
    df_false_positives = pd.DataFrame({'Image':imagelist, 'Label':labellist})
    df_false_positives.to_csv('labels.csv', index=False) 
    print('\n\nfalse_positives dataframe written to labels.csv')
    df_false_positives
    return df_false_positives

#tag_false_positives()
```

# MAIN

```python
dfbb = get_bounding_boxes(f'{DATADIR}/{CVATXMLFILE}')

# Extract random image samples for each damage level label and write these to jpg image files.
        
cap = cv2.VideoCapture(f'{DATADIR}/{VIDEOFILE}')
grouped = dfbb.groupby('label')
print(grouped.size())
for name, df in grouped:
    n = min(int(NSAMPLES), df.shape[0]) # get min of row count or NSAMPLES 
    for i, r in df.sample(n).iterrows():
        cap.set(1, r.frame)
        _, frame = cap.read()
        img = frame[r.ytl:r.ybr, r.xtl:r.xbr]
        cv2.imwrite(f'{r.label}-t{r.track_id:04d}-f{r.frame:04d}.jpg', img)
cap.release()

imagelist = glob.glob('*.jpg')
np.random.shuffle(imagelist)
df = pd.DataFrame(imagelist,columns=['image'])
df.image = df.apply(lambda r: f'{r.image}', axis=1)

tag_false_positives()
```
